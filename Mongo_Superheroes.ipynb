# Query the SQLite database using the sqlite package

```python
#connect to the data base 
files = "superheroes.db"
conn = sqlite3.connect(files)
cursor = conn.cursor()
```

```python
cursor.execute("""
SELECT
      title, s.name, r.name
  FROM 
      Movies AS m
  INNER JOIN 
       Superheroes AS s
    ON
      m.superhero_id = s.id 
  INNER JOIN
      RealNames AS r
    ON 
      r.superhero_id = s.id 
""")
print(cursor.fetchall())
```
